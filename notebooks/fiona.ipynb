```python
import os
import geopandas as gpd

def save_csvs_to_geopackage(csv_directory, output_gpkg, lat_col="Start_Lat", lon_col="Start_Lng", crs="EPSG:4326"):
    """
    Convert all CSV files in a directory to layers in a single GeoPackage.
    
    Parameters:
        directory (str): The path to the directory containing the CSV files.
        output_gpkg (str): The path to the output GeoPackage file.
        lat_col (str): Name of the column containing latitude values (default: "Latitude").
        lon_col (str): Name of the column containing longitude values (default: "Longitude").
        crs (str): The CRS to use for the GeoDataFrame (default: "EPSG:4326").
    """
    # Loop through all files in the directory
    for filename in os.listdir(csv_directory):
        if filename.endswith(".csv"):
            csv_file = os.path.join(csv_directory, filename)
            layer_name = os.path.splitext(filename)[0]  # Use the filename (without extension) as the layer name
            
            # Convert CSV to GeoDataFrame
            gdf = gpd.read_file(csv_file)
            gdf['geometry'] = gpd.points_from_xy(gdf[lon_col], gdf[lat_col])
            print(type(gdf))
            # gdf = gdf.set_crs(crs, inplace=True)
            
            # # Save GeoDataFrame as a layer in the GeoPackage
            # gdf.to_file(output_gpkg, layer=layer_name, driver="GPKG", append=True)
            # print(f"Layer '{layer_name}' saved to {output_gpkg}.")
```
